# 1. Calculate the total number of employees in each department and from each location.

```python
print(data3.groupby("Training Program Name").size())
```

```python
data3.groupby("Training Program Name").size().plot(kind="bar",color="black")
plt.title("No of employees in each department")
plt.xlabel("Name of the department")
plt.ylabel("No of employees")
plt.show()
```

```python
print(data3.groupby("Location").size())
```

```python
data3.groupby("Location").size().plot(kind="bar",color="black")
plt.title("No of employees in each Location")


plt.show()
```
